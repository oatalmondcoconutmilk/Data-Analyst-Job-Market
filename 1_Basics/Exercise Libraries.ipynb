```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

location_counts = df['job_country'].value_counts().head(5)
plt.bar(location_counts.index, location_counts.values, color='orange')
plt.title('Number of Jobs per Country')
plt.xlabel('Country')
plt.ylabel('Number of Postings')
plt.xticks(rotation = 45, ha = "right")
plt.show()
```

```python

avg_salary = df.groupby('job_title_short')['salary_year_avg'].mean().sort_values(ascending=True)

plt.barh(avg_salary.index, avg_salary.values)
plt.xlabel('Average Salary')
plt.ylabel('Average Salary by Job Title')
plt.xticks(rotation = 45, ha = "right")
plt.show()
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['month'] = df['job_posted_date'].dt.month
df_filteres = df[df['month'] <= 6]
avg_salary_per_month = df_filteres.groupby('month')['salary_year_avg'].mean()
plt.plot(avg_salary_per_month.index, avg_salary_per_month.values, color = 'red',label='Average Salary')
plt.xlabel('Month')
plt.ylabel('Average Salary')
plt.title('Average Salary by Job Month')
plt.legend()
plt.show()
```

```python
top_5_jobs = df['job_via'].value_counts().head(5)
top_5_jobs.plot(kind = 'bar')
plt.xlabel(' ')
plt.xticks(rotation = 45, ha = "right")
```

```python
import pandas as pd

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['month_name'] = df['job_posted_date'].dt.to_period('M')

monthly_job_counts = df['month_name'].value_counts().sort_index()
monthly_job_counts.plot( title = 'Trend of Job Postings Over Time')
plt.xlabel('Month')
plt.ylabel('number of Job Postings')
plt.legend()
plt.show()
```

```python
top_locations = df['job_location'].value_counts().head(5).sort_values(ascending=True)
top_locations.plot(kind = "barh")
plt.xlabel('Count of Jobs')
plt.ylabel('Job Location')
plt.title('Number of Job Postings by Job Location')
plt.show()
```
